LIFESTYLE RISK CLUSTERING

```python

# Create risk score based on multiple health factors
df['Risk_Score'] = (
    (df['BMI'] > 30).astype(int) +
    (df['Exercise_Freq'] == 'None').astype(int) +
    (df['Diet_Quality'] == 'Poor').astype(int) +
    (df['Alcohol_Consumption'] == 'High').astype(int) +
    (df['Stress_Level'] > 7).astype(int) +
    (df['Sleep_Hours'] < 6).astype(int)
)

# Assign risk level based on score
df['Risk_Level'] = pd.cut(
    df['Risk_Score'],
    bins=[-1, 1, 3, 6],
    labels=['Low', 'Medium', 'High']
)

# Plot the pie chart
risk_counts = df['Risk_Level'].value_counts()
colors = ['green', 'yellow', 'red']

plt.figure(figsize=(5,5))
plt.pie(risk_counts, labels=risk_counts.index, autopct='%1.1f%%',  colors=colors)
plt.title('Distribution of Lifestyle Risk Levels')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
